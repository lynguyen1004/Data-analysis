```python
import time
def times(x):
    now = time.gmtime(int(x))
    result='{}-{}-{}'.format(now.tm_year, now.tm_mon, now.tm_mday)
    return result
```

```python
k1['time'].apply(times)
```

```python
k1['times']=k1['time'].apply(times)
k1
```
